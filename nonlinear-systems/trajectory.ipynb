# Plotting Trajectories of Second Order Nonlinear Systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Limits of Plot

```python
X_MIN = -2
X_MAX = 2

Y_MIN = -2
Y_MAX = 2

X_STEP = 0.2
Y_STEP = 0.2
```

```python
x = np.arange(X_MIN, X_MAX+X_STEP, X_STEP)
y = np.arange(Y_MIN, Y_MAX+Y_STEP, Y_STEP)
x1, x2 = np.meshgrid(x, y)

plt.plot(x1,x2, marker='.', color='k', linestyle='none')
plt.title("Plotting Region")
plt.show()
```

## Dynamic Equation of system

```python
dx1 = x2 + x1*(1 - x1*x1 - x2*x2)
dx2 = -x1 + x2*(1 - x1*x1 - x2*x2)
```

### Normalize vectors prior to plotting

```python
norm = np.sqrt(dx1*dx1+dx2*dx2)
norm[norm==0] = 1
ndx1 = dx1/norm
ndx2 = dx2/norm
```

```python
plt.quiver(x1, x2, ndx1, ndx2)
plt.title("Trajectories")
plt.show()
```
